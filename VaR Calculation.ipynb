```python
# This function calculates the 1 day shift, i.e. exp(ln(FXrate1/FXrate2))-1 (for a list of FX rates).
# The function returns a list of 1 day shift values, i.e:
# for a list containing [FXrate1, FXrate2, FXrate3, ....]
# it calculates the 1 day shift values and returns them in a list as follows: [exp(ln(FXrate1/FXrate2))-1 , exp(ln(FXrate2/FXrate3))-1, ... ]
def CalcOneDayShiftForFX(mrList):
    try:
        # with zip(mrList,mrList[1:]) we can get value pairs from the input mrList ([FXrate1, FXrate2, FXrate3, ....], like:
        # (FXrate1, FXrate2) and (FXrate2, FXrate3), and apply a division operation on each value pair.
        xDivy = [x/y for x,y in zip(mrList,mrList[1:])]
        # we calculate exp(ln(z))-1 for each z from the obtained x/y values from the previous step
        return list(map(lambda z: math.exp(math.log(z))-1,xDivy))
    except Exception as e:
        print("Exception occured in function CalcOneDayShiftForFX: ",e)
```

```python
# this function calculates and returns the 1 day VaR value
# with as input 2 lists of FX rates fxRateList1,fxRateList2 and 
# 2 spot portfolio values for CCY1 and CCY2
def CalcVaROneDay(fxRateList1, fxRateList2,spotPortfolioValueCCY1,spotPortfolioValueCCY2):
    try:
        # calculate the 1 day shift values for the list of FX rates for CCY1
        oneDayShiftCCY1 = CalcOneDayShiftForFX(fxRateList1)
        # for illustration and checking the calculated values the last 5 values from the list will be printed
        print("1 day shift for CCY 1 (last 5 values from the list):",oneDayShiftCCY1[-5:])

        # Calculate PnL Vector for CCY1 = spotPortfolioValue * oneDayShift value
        pnlVectorCCY1 = [spotPortfolioValueCCY1 * el for el in oneDayShiftCCY1]
        print("PnL Vector for CCY 1 (last 5 values from the list):",pnlVectorCCY1[-5:])

        # calculate the 1 day shift values for the list of FX rates for CCY2
        oneDayShiftCCY2 = CalcOneDayShiftForFX(fxRateList2)
        print("1 day shift for CCY 2 (last 5 values from the list):",oneDayShiftCCY2[-5:])

        # Calculate PnL Vector for CCY2 = spotPortfolioValue * oneDayShift value
        pnlVectorCCY2 = [spotPortfolioValueCCY2 * el for el in oneDayShiftCCY2]
        print("PnL Vector for CCY 2 (last 5 values from the list):",pnlVectorCCY2[-5:])

        # total PnL = PnLVectorCCY1 + PnLVectorCCY2
        totalPnl = [x+y for x,y in zip(pnlVectorCCY1,pnlVectorCCY2)]
        print("Total PnL (last 5 values from the list):",totalPnl[-5:])
        print()
        
        # then we sort the list of PnL values in ascending order
        sortedTotalPnl = sorted(totalPnl)
        # the second worst PnL value will then be second item in the sorted list
        secondWorstPnL = sortedTotalPnl[1]
        # the third worst PnL value will then be third item in the sorted list
        thirdWorstPnL = sortedTotalPnl[2]
        # A VaR with a confidence level of 99% for this FX portfolio example
        # will be based on 40% of the second worst PnL and 60% of the third worst PnL
        # finally return the calculation 0.4*secondWorstPnL+0.6*thirdWorstPnL
        return 0.4*secondWorstPnL+0.6*thirdWorstPnL
    except Exception as e:
        print("Exception occured in function CalcVaROneDay: ",e)
```
